```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x= np.linspace(0, 10*np.pi, 1000)
y = np.sin(x)
```

```python
print(x[:10])
print(y[:10])
```

```python
plt.plot(x, y)
plt.title("Sine wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
